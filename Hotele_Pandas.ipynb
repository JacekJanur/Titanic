(porządkowe to takie dzięki któremu można uporządkować dane, np. ilość domów a dane nominalne to, np. kolor ścian domu). <br />
Napewno musimy wykorzystać technikę "gorącojedynkową" dla kolumn takich jak "reserved_room_type", aby przyszłe algorytmy mogły je zinterpretować, oraz nie uważały, że pokój typu A jest większy od pokoju typu C

```python
df.select_dtypes("object")
```

Teraz czas się zastanowić, jaki typ danych posiada dana kolumna, np. hotel to napewno jest dana nominalna a miesiąc możemy traktować jak daną porządkową. Kolumny gorącojedynkowe możemy uzyskać w następujący sposób:

```python
pd.get_dummies(df[['hotel', 'reserved_room_type', 'assigned_room_type', 'deposit_type']])
```

Jako przykład danych porządkowych spróbujemy zastosować kolumnę "meal", ponieważ można to potraktować myśleniem "tamten klient miał bardziej eksluzywną usługę". Danymi porządkowymi możemy poradzić sobie następująco:

```python
df["meal"] = df["meal"].astype('category')
df.dtypes
```

```python
df["meal_cat"] = df["meal"].cat.codes
df[['meal','meal_cat']]
```
